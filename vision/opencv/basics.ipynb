```python
import cv2
import imutils as imu
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
im = cv2.imread('utils/jp.png')
(h,w,d) = im.shape
print(im.shape)
def show(im):
    plt.imshow(im[:,:,::-1])
show(im)
```

```python
show(im[60:140,440:500])
# [upper:lower,left:right]
```

```python
r = cv2.resize(im, (200,100))
print(r.shape, end=' ')
# or if we dont know 100
r = imu.resize(im, width=200)
print(r.shape)
show(r)
```

```python
center = ( w//2, h//2 )
M = cv2.getRotationMatrix2D(center, -45, 1.0)
r = cv2.warpAffine(im, M, (w,h))
# or
r = imu.rotate(im, -45)
show(r)
```

```python
r = imu.rotate_bound(im, -45)
show(r)
```

```python
r = cv2.GaussianBlur(im, (11,11), 0)
show(r)
```

```python
r = im.copy()
ul_x, ul_y = 400, 100 # upper left
lr_x, lr_y = 500, 200 # lower right
cv2.rectangle(r, (ul_x,ul_y), (lr_x,lr_y), (0,255,255), 2)
show(r)
```

```python
r = im.copy()
x, y = 500, 100
cv2.circle(r, (x,y), 30, (255,255,255), -1)
show(r)
```

```python
r = im.copy()
ul_x, ul_y = 400, 100 # upper left
lr_x, lr_y = 500, 200 # lower right
cv2.line(r, (ul_x,ul_y), (lr_x,lr_y), (0,255,255), 5)
show(r)
```

```python
r = im.copy()
x, y = 10, 25
font, font_size, boldness = cv2.FONT_HERSHEY_SIMPLEX, 0.7, 2
cv2.putText(r, "Testing...", (x,y), font, font_size, (0, 255, 255), boldness)
show(r)
```
