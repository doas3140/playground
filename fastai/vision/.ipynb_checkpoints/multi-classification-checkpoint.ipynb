```python
%reload_ext autoreload
%autoreload 2
%matplotlib inline

from fastai.vision import *
from fastai.metrics import error_rate

from preparation import get_logos, get_nr2classname_dict
nr2c = get_nr2classname_dict()
c2nr = {a:int(aa) for aa,a in nr2c.items()}
logos = get_logos()
```

```python
multiclass = []

for l in logos:
    c = ''
    for a in l.ann:
        if a.c not in c:
            if c != '':
                c += ' '
            c += a.c
    multiclass.append(c)
multiclass[:5]

name2c = {x.path.name:c for x,c in zip(logos,multiclass)}
_ = [print(x, end='\n') for x in set(name2c.values())]
```

```python
files = [x.path for x in logos if x.logo == True]
test_idx = [i for i,x in enumerate(logos) if x.test]
name2test = { x.path.name:x.test for x in logos } # is logo?

tfms = get_transforms(do_flip=False, max_rotate=180, max_zoom=1.5)

data = (ImageItemList(files)
            .split_by_valid_func(lambda x: name2test[ x.name ])
            .label_from_func(lambda x: name2c[ x.name ], label_delim=' ')
            .transform(tfms, size=224)
            .databunch(bs=64, num_workers=4)
            .normalize(imagenet_stats)
       )
```

```python
data.show_batch()
```

```python
valid_dl = data.dl(ds_type=DatasetType.Valid)
for i,(x,y) in enumerate(valid_dl):
    print(x.shape, y.shape, y[0][30:35].int())
    if i > 0: break
```

```python
def custom_acc(y_pred, y_true, t = 0.2):
    return ((y_pred > t) == y_true.byte()).float().mean()

acc = partial(accuracy_thresh, thresh=0.2) # same as custom w/ fastai
f2_score = partial(fbeta, thresh=0.2)
learn = create_cnn(data, models.resnet34, metrics=[custom_acc, f2_score])
```

```python
learn.fit_one_cycle(4)
```

```python
learn.lr_find()
learn.recorder.plot()
```

```python
learn.fit_one_cycle(2, 1e-2)
```

```python
learn.unfreeze()
learn.lr_find(start_lr=1e-8)
learn.recorder.plot()
```

```python
learn.fit_one_cycle(3, max_lr=slice(1e-4,1e-3))
```

```python
interp = ClassificationInterpretation.from_learner(learn)
```

```python
interp.plot_multi_top_losses()
```
